Importing The dependencies

```python
import numpy as np
import pandas as pd  #dataframe
import matplotlib.pyplot as plt
import seaborn as sns  #visualize random distributonss
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor   #uses gradient boost trees
from sklearn import metrics
```

Importing the California House price dataset

```python
house_price_dataset=sklearn.datasets.fetch_california_housing()
```

```python
print(house_price_dataset)
```

```python
# Loading the dataset to the pandas dataframe
house_price_dataframe=pd.DataFrame(house_price_dataset.data,columns=house_price_dataset.feature_names)
```

```python
house_price_dataframe.head()
```

```python
#Add the target("price") clumn to the dataset
house_price_dataframe["price"]=house_price_dataset.target
```

```python
house_price_dataframe.head()
```

```python
#chcking number of rows and columns we have in dataframe
house_price_dataframe.shape
```

```python
#Check for missing values
house_price_dataframe.isnull().sum()
```

```python
#statistical measures of dataset like mean medain and that all
house_price_dataframe.describe()
```

Understnd the correlation bw features of dataset

1.Positive Correlation (if one row increases the other also increases)
2.Negative Correlation (if one row decreases the other also decreases)


```python
correlation=house_price_dataframe.corr()
```

```python
#Construct heat map for correlation
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True,square=True,fmt="0.1f",annot=True,annot_kws={"size":8},cmap="Blues")
```

Splitting the Data abd the target

```python
X=house_price_dataframe.drop(["price"],axis=1)
Y=house_price_dataframe["price"]
```

```python
print(X)
print(Y)
```

Splitting data into trAining and test data

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

TRAINING OUR MODEL USING XGBOOST (decsion model)

```python
#Load the model
model=XGBRegressor()
```

```python
#train the model with X_train
model.fit(X_train,Y_train)
```

Evaluation

Prediction of rprice of the houses

```python
#Accuracy for prediction on training data
training_data_prediction=model.predict(X_train)
```

```python
print(training_data_prediction)
```

```python
#R squared Error   (For mean and median)
score_1=metrics.r2_score(Y_train,training_data_prediction) #Finds the variance bw these two (the r2_score)

#Mean Absolute Error
score_2=metrics.mean_absolute_error(Y_train,training_data_prediction)

print("R squared Error: " ,score_1) #Lesser thhe value more accuracy the model has close to 1 it should be
print("Mean absolute Error: " ,score_2)
```

Visualizing the actual Prices and predicted prices

```python
plt.scatter(Y_train,training_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual Price VS Predicted Price")
plt.show()       #See values are very close so our model is correct for now
```

Prediction oN test Data as well

```python
#Accuracy for prediction on test data
test_data_prediction=model.predict(X_test)
```

```python
print(test_data_prediction)
```

```python
#R squared Error   (For mean and median)
score_1=metrics.r2_score(Y_test,test_data_prediction) #Finds the variance bw these two (the r2_score)

#Mean Absolute Error
score_2=metrics.mean_absolute_error(Y_test,test_data_prediction)

print("R squared Error: " ,score_1) #Lesser thhe value more accuracy the model has close to 1 it should be
print("Mean absolute Error: " ,score_2)
```
